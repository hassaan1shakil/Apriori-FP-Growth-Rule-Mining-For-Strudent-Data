```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('DATA.csv')
df.head(1)
```

```python
data = df[['Graduated high-school type', 'Scholarship type', 'Additional work', 'Regular artistic or sports activity', 'Transportation to the university', 'Accommodation', "Mother education", ' Fathers education', 'Parental status', "Mothers occupation", "Father occupation:", 'Weekly study hours', 'Attendance to classes', 'CGPA last sem']].copy()
data
```

```python
# Preprocessing

scholarship_map = {1: 'None', 2: 'Low', 3: 'Medium', 4: 'High', 5: 'Full'}
work_map = {1: 'Work', 2: 'No Work'}
activities_map = {1: 'Active', 2: 'Inactive'}
transporatation_map = {1: 'Bus', 2: 'Private', 3: 'Bicycle', 4: 'Other Transport'}
accomodation_map = {1: 'Rental', 2: 'Dorm', 3: 'Home', 4: 'Other Accomodation'}
mother_edu_map = {1: 'm_edu_primary', 2: 'm_edu_secondary', 3: 'm_edu_high', 4: 'm_edu_uni', 5: 'm_edu_masters', 6: 'm_edu_phd'}
father_edu_map = {1: 'f_edu_primary', 2: 'f_edu_secondary', 3: 'f_edu_high', 4: 'f_edu_uni', 5: 'f_edu_masters', 6: 'f_edu_phd'}
parental_status_map = {1: 'married', 2: 'divorced', 3: 'died'}
mother_job_map = {1: 'mother_retired', 2: 'housewife', 3: 'mother_government officer', 4: 'mother_private sector employee', 5: 'mother_self-employment', 6: 'mother_other_job'}
father_job_map = {1: 'father_retired', 2: 'father_government officer', 3: 'father_private sector employee', 4: 'father_self-employment', 5: 'father_other_job'}
study_map = {1: 'No Study', 2: 'Less than 5 hours', 3: '6-10 hours', 4: '11-20 hours', 5: 'More than 20 hours'}
attendance_map = {1: 'Present', 2: 'Partially Present', 3: 'Absent'}
cgpa_map = {1: 'Warning', 2: 'Below Average', 3: 'Average', 4: 'Above Average', 5: 'Exceptional'}
```

```python
# Updating Mappings of Attributes

data['Scholarship type'] = df['Scholarship type'].map(scholarship_map)
data['Additional work'] = df['Additional work'].map(work_map)
data['Regular artistic or sports activity'] = df['Regular artistic or sports activity'].map(activities_map)
data['Transportation to the university'] = df['Transportation to the university'].map(transporatation_map)
data['Accommodation'] = df['Accommodation'].map(accomodation_map)
data['Mother education'] = df['Mother education'].map(mother_edu_map)
data[' Fathers education'] = df[' Fathers education'].map(father_edu_map)
data['Parental status'] = df['Parental status'].map(parental_status_map)
data['Mothers occupation'] = df['Mothers occupation'].map(mother_job_map)
data['Father occupation:'] = df['Father occupation:'].map(father_job_map)
data['Weekly study hours'] = df['Weekly study hours'].map(study_map)
data['Attendance to classes'] = df['Attendance to classes'].map(attendance_map)
data['CGPA last sem'] = df['CGPA last sem'].map(cgpa_map)
```

```python
data.head()
```

```python
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

duplicates = data.duplicated().sum()
print("\nNumber of Duplicate Rows:", duplicates)

summary = data.describe()
print("\nSummary Statistics:\n", summary)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth, association_rules

transactions = data.astype(str).values.tolist()
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
data_itemsets = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Support & Confidence # 1

frequent_itemsets = apriori(data_itemsets, min_support=0.3, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4)
top_rules = rules.sort_values(by='confidence', ascending=False).head(10)
top_rules
```

```python
# Support & Confidence # 2

frequent_itemsets = apriori(data_itemsets, min_support=0.4, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
top_rules = rules.sort_values(by='confidence', ascending=False).head(10)
top_rules
```
